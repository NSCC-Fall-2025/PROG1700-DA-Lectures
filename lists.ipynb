```python
# basic list   0       1       2        3        4
usernames = ["john","jacob","jingle","hymer","schmitt", "jill"]
print(usernames)

# "5" is a "magic number"
for i in range(5): # 0 1 2 3 4
    print(usernames[i])
```

```python
# how many items are in the list?
usernames = ["john","jacob","jingle","hymer","schmitt", "jill", "joe"]
len(usernames)
```

```python
# iterating through list
usernames = ["john","jacob","jingle","hymer","schmitt", "jill"]

# one way, works but wordy
for i in range(len(usernames)):
    print(usernames[i])
print("--")
# also works
i = 0
while (i < len(usernames)):
    print(usernames[i])
    i += 1

print("--")
# a list specific way (shorter and easier to read/use)
for username in usernames:
    print(username)
```

```python
# what if I don't know what the items are?
data = [0] * 100
i = 0
while True:
    user_input = input("Enter a number: ")
    if user_input == "":
        break
    data[i] = user_input
    i += 1
    if i >= len(data):
        break

print(data)
```
