```python
# a really big if statement with all the fixings
grade = 34

if grade > 90:
    print("A")
elif grade > 10:
    print("possibly an F?")
    if grade >= 60:
        print("Not an F")
    elif grade > 40:
        print("Not doing horribly....")
        if grade > 50:
            print("Well, better than a 40!")
        else:
            print("Need a little a more effort..")
    else:
        print("Better luck next time.")
else:
    print("Not an A")
```

```python
# program to mix paint colours

# given two primary colours
colour1 = "blue"
colour2 = "red"

# determine the mixed colour
if colour1 == "blue":
    if colour2 == "yellow":
        print("green")
    elif colour2 == "red":
        print("violet")
elif colour1 == "red":
    if colour2 == "yellow":
        print("orange")
    elif colour2 == "blue":
        print("violet")

# swap colours if needed
if colour2 == "blue" or colour1 == "yellow":
    temp = colour1
    colour1 = colour2
    colour2 = temp
        
if colour1 == "blue" and colour2 == "yellow":
    print("green")
elif colour1 == "red" and colour2 == "yellow":
    print("orange")
elif colour1 == "blue" and colour2 == "red":
    print("violet")
    
```
