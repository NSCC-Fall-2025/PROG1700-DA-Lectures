```python
# high / low game
import random

def generate_answer():
    # generate a random number between 1 - 10
    answer = random.randint(1, 10)
    print(answer)
    return answer

def take_a_guess():
    # ask user to make a guess
    guess = int(input("Guess a number: "))
    return guess

def check_answer(answer, guess):
    if answer < guess:
        print("Too High")
        return False
    elif answer > guess:
        print("Too Low")
        return False
    else:
        return True
    
def check_for_winner(answer, guess):
    if answer == guess:
        print("You WIN!")
    else:
        print("You LOSE!")

def high_low_game():
    allowed_guesses = 3
    # create the random number    
    answer = generate_answer()
    # game loop
    num_guesses = 0
    while num_guesses < allowed_guesses:
        guess = take_a_guess()
        win = check_answer(answer, guess)
        if win:
            break
        num_guesses += 1

    check_for_winner(answer, guess)

# start game
high_low_game()
```
