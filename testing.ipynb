```python
import ipytest, pytest
ipytest.autoconfig()
```

```python
# a program that adds two numbers

def add(a, b):
    return a + b

# unit test to determine if program works
assert add(2, 3) == 5
```

```python
# create a program that asks for the user's age
# the program will return true, if the user is an adult

def get_user_age():
    return int(input("Enter age: ")) >= 18

assert get_user_age() == True
```

```python
# pytest basics
%%ipytest

# functions start with "test_"
# contains asserts that are testing a single aspect of the program
def test_number_assumptions():
    age = 15

    # check to see if the age is a particular value
    assert age == 25

    # check to see if the age is over 18
    assert age > 18

    # check to see if the age is under 100
    assert age < 100
```

```python
# code to test

def get_list_of_colours():
    return ["red", "blue", "green"]
```

```python
%%ipytest
# pytest functions

def test_list_contents():
    colors = get_list_of_colours()

    assert colors[0] == "red"
```

```python
%%ipytest

# special asserts
def test_special_asserts():
    assert 1 < 2, "1 should be less than 2"
    assert 2.45 > 1.23, "big decimals are greater than little decimals"
    assert len("Hello World!") > 0, "the string's length should be > 0"
    
    my_dict = {
        "name": "John",
        "id": "W213456"
    }
    assert my_dict["name"] == "John", "John is my name too...."
    assert "W" in my_dict["id"], "the id should have a W"
    assert "id" in my_dict, "the info should have an id"
```

```python
%%ipytest

# mocking
def test_mocking(mocker):
    mocker.patch('builtins.input', return_value="Goodbye World!")

    user_input = input("Enter something: ")

    assert(user_input == "Hello World!")

    print(f"You typed: {user_input}")
```
